# Q7.Go to webpage https://www.amazon.in/ Enter “Laptop” in the search field and then click the search icon. Then set CPU Type filter to “Intel Core i7” as shown in the below image:

After setting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
1. Title
2. Ratings
3. Price

```python
# Opening the naukri page on automated web browser
driver.get("https://www.amazon.in/")
```

```python
# Find the search field element
designation = driver.find_element(By.ID, 'twotabsearchtextbox')

# Enter the desired designation
designation.send_keys("Laptop")
```

```python
# Click the search icon
search_icon = driver.find_element(By.ID, "nav-search-submit-button")
search_icon.click()
```

```python
# Wait for the first 10 laptops to be loaded
laptops = WebDriverWait(driver, 10).until(EC.visibility_of_all_elements_located((By.XPATH, '//div[@data-component-type="s-search-result"]')))[:10]

# Create an empty list to store the laptop data
laptop_data = []

# Iterate over the laptops and scrape the desired attributes
for laptop in laptops:
    # Find the title
    title_element = laptop.find_element(By.XPATH, './/h2/a')
    title = title_element.get_attribute("innerHTML").strip()
    
    # Find the rating
    try:
        rating_element = laptop.find_element(By.XPATH, './/span[@class="a-icon-alt"]')
        rating = rating_element.get_attribute("innerHTML")
    except:
        rating = "N/A"
    
    # Find the price
    try:
        price_element = laptop.find_element(By.XPATH, './/span[@class="a-price-whole"]')
        price = price_element.get_attribute("innerHTML")
    except:
        price = "N/A"
    
    # Append the laptop data to the list
    laptop_data.append({"Title": title, "Rating": rating, "Price": price})

# Create a DataFrame from the laptop data
df = pd.DataFrame(laptop_data)

# Set the display options to show all rows and columns
pd.set_option('display.max_colwidth', None)

# Print the DataFrame
df
```
